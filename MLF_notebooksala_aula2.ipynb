```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans, DBSCAN
from sklearn.datasets import load_diabetes, make_blobs, make_moons, make_circles
from sklearn.linear_model import LinearRegression
from sklearn.metrics import root_mean_squared_error, silhouette_score, calinski_harabasz_score
from sklearn.model_selection import train_test_split, cross_val_score, cross_validate
from sklearn.mixture import GaussianMixture
from sklearn.tree import DecisionTreeRegressor, plot_tree
```

# Clusterização

## Dados agrupados em circulos

```python
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.3, random_state=0)

plt.scatter(X[:,0], X[:,1])

plt.show()
```

### K-means

```python
kmeans = KMeans(n_clusters=4, init='k-means++', n_init=10, max_iter=300, tol=1**-4, random_state=0)

pred_y = kmeans.fit_predict(X)

plt.scatter(X[:,0], X[:,1])
plt.scatter(X[:, 0], X[:, 1], c=pred_y)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red')


score_silhouette = silhouette_score(X, pred_y)
score_ch = calinski_harabasz_score(X, pred_y)

print(f"score silhueta {score_silhouette:.2f}, Calinski and Harabasz {score_ch:.2f}")


plt.show()
```

### DBSCAN (Modelo densidade)

```python
dbscan = DBSCAN(eps=1, min_samples=20)

pred_y = dbscan.fit_predict(X)

plt.scatter(X[:,0], X[:,1])
plt.scatter(X[:, 0], X[:, 1], c=pred_y)

score_silhouette = silhouette_score(X, pred_y)
score_ch = calinski_harabasz_score(X, pred_y)

print(f"score silhueta {score_silhouette:.2f}, Calinski and Harabasz {score_ch:.2f}")

plt.show()
```

### Gaussian Mixture Models (Baseado em distribuição)

```python
gmm = GaussianMixture(n_components=4, covariance_type="full", random_state=42)

pred_y = gmm.fit_predict(X)

plt.scatter(X[:,0], X[:,1])
plt.scatter(X[:, 0], X[:, 1], c=pred_y)

score_silhouette = silhouette_score(X, pred_y)
score_ch = calinski_harabasz_score(X, pred_y)

print(f"score silhueta {score_silhouette:.2f}, Calinski and Harabasz {score_ch:.2f}")

plt.show()
```

# Regressão

## Predizendo evolução de diabetes

[Link para dataset utilizado](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_diabetes.html)

### Aplicação de árvores de decisão

```python
X_diabetes, y_diabetes = load_diabetes(return_X_y=True)

X_diabetes_train, X_diabetes_test, y_diabetes_train, y_diabetes_test = train_test_split(X_diabetes, y_diabetes, test_size=0.3)

regressor = DecisionTreeRegressor()

regressor.fit(X_diabetes_train, y_diabetes_train)
```

```python
_ = plot_tree(regressor)
```

### Aplicação de regressão linear

```python
regressor_linear = LinearRegression()

regressor_linear.fit(X_diabetes_train, y_diabetes_train)
```

### Comparando as soluções

```python
y_pred_arvore_regressao = regressor.predict(X_diabetes_train)
y_pred_regressao_linear = regressor_linear.predict(X_diabetes_train)

RMSE_arvore_regressao = root_mean_squared_error(y_diabetes_train, y_pred_arvore_regressao)
RMSE_regressao_linear = root_mean_squared_error(y_diabetes_train, y_pred_regressao_linear)

print(f"RMSE da arvore de regressao eh {RMSE_arvore_regressao:.2f}, enquanto da regressao linear {RMSE_regressao_linear:.2f}")
```

```python
y_pred_arvore_regressao = regressor.predict(X_diabetes_test)
y_pred_regressao_linear = regressor_linear.predict(X_diabetes_test)

RMSE_arvore_regressao = root_mean_squared_error(y_diabetes_test, y_pred_arvore_regressao)
RMSE_regressao_linear = root_mean_squared_error(y_diabetes_test, y_pred_regressao_linear)

print(f"RMSE da arvore de regressao eh {RMSE_arvore_regressao:.2f}, enquanto da regressao linear {RMSE_regressao_linear:.2f}")
```

### Utilizando validação cruzada (cross validation)

```python
regressor_linear_cross = LinearRegression()

scores = cross_val_score(regressor_linear_cross, X_diabetes, y_diabetes, cv=5, scoring="neg_root_mean_squared_error")
```

```python
print(f"O RMSE médio foi de {-scores.mean()}")
```

### Como descobrir o melhor modelo?

```python
regressor_linear_cross = LinearRegression()

results = cross_validate(regressor_linear_cross, X_diabetes, y_diabetes, cv=5, scoring="neg_root_mean_squared_error", return_estimator=True)
```
